# Height and Weight

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
HW = pd.read_csv("./data/HWdata.csv", header=None, names=['Height', 'Weight'])
HW
```

The data frame contains the height and weight data. Heights are in the first column, and weights are in the second.

```python
plt.figure()
plt.scatter(HW['Height'], HW['Weight'])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

The height column of HW has the heights in cm in increasing order.

```python
plt.figure()
plt.scatter(HW['Height'][0:10], HW['Weight'][0:10])
plt.scatter(HW['Height'][10:], HW['Weight'][10:])
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
