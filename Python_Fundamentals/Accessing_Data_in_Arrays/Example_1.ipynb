# Gas Prices in the 1990's vs 2000's

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
```

```python
gas_prices = pd.read_csv("./data/gasprices.csv", header=4)
gas_prices.head()
```

```python
Year = gas_prices['Year']
France = gas_prices['France']
Germany = gas_prices['Germany']

plt.figure()
plt.plot(Year, France)
plt.plot(Year, Germany)
plt.title("Gasoline Prices in USD/gal over time")
plt.xticks([y for y in range(Year.min(), Year.max() + 2, 2)])
plt.legend(["France", "Germany"])
plt.show()
```

Create a vector of data values for France and Germany for the 1990s.

```python
France_90s = France[:10]
France_90s
```

```python
Germany_90s = Germany[:10]
Germany_90s
```

```python
Year_90s = Year[:10]
Year_90s
```

Let's plot the data.

```python
plt.figure()
plt.plot(Year_90s, France_90s)
plt.plot(Year_90s, Germany_90s)
plt.title("Gasoline Prices in USD/gal in the 90s")
plt.legend(["France", "Germany"])
plt.xticks(Year_90s)
plt.show()
```
