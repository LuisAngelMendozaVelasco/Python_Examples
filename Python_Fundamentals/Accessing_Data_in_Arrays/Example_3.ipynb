# Indexing into Matrices with Vectors

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
```

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None).to_numpy()
year = pd.read_excel("./data/electricity.xlsx", sheet_name='year', header=None).to_numpy()
```

```python
plt.figure()
plt.plot(year.T, price.T)
plt.title("Monthly Electricity Prices")
plt.legend(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```

Electricity price data from January 1990 to December 2019 are stored in the array price. Each column is a year of monthly prices. Let's extract the prices form January.

```python
january = price[0, :]
pd.Series(january, index=year[0], name='January')
```

We can also extract multiple rows or columns.

```python
quarters = price[[0, 3, 6, 9]]
pd.DataFrame(quarters, index=['January', 'April', 'July', 'October'], columns=year[0])
```

We can see in the plot that there are different trends in the 1990's and 2000's.

```python
the_90s = quarters[:, :10]
pd.DataFrame(the_90s, index=['January', 'April', 'July', 'October'], columns=year[0, :10])
```

```python
the_2000s = quarters[:, 10:]
pd.DataFrame(the_2000s, index=['January', 'April', 'July', 'October'], columns=year[0, 10:])
```
