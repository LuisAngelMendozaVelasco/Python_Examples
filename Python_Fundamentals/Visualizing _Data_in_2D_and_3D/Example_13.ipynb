# Electricity Revenue and Usage

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)
usage = pd.read_excel("./data/electricity.xlsx", sheet_name='usage', header=None)
year = pd.read_excel("./data/electricity.xlsx", sheet_name='yr', header=None)
month = pd.read_excel("./data/electricity.xlsx", sheet_name='mth', header=None)

revenue = price * usage
```

We can use the [plot_surface()](https://matplotlib.org/stable/gallery/mplot3d/surface3d.html#d-surface-colormap) function to visualize an m-by-n matrix by representing each value in the matrix as the height of a surface.

```python
X = np.arange(0, revenue.shape[0])
Y = np.arange(0, revenue.shape[1])
X, Y = np.meshgrid(X, Y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, revenue.T, cmap=cm.coolwarm)
ax.set_xlabel("Month")
ax.set_ylabel("Year")
ax.set_zlabel("Revenue")
plt.show()
```

Let's pass the months and years as X, Y inputs.

```python
X, Y = np.meshgrid(month.T, year)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, revenue.T, cmap=cm.coolwarm)
ax.set_xlabel("Month")
ax.set_ylabel("Year")
ax.set_zlabel("Revenue")
plt.show()
```

You can use the same syntax with the [plot_wireframe()](https://matplotlib.org/stable/gallery/mplot3d/wire3d.html#d-wireframe-plot) function.

```python
X, Y = np.meshgrid(month.T, year)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_wireframe(X, Y, revenue.T)
ax.set_xlabel("Month")
ax.set_ylabel("Year")
ax.set_zlabel("Revenue")
plt.show()
```

You can use a [contour](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contour.html#matplotlib-pyplot-contour) plot to visualize matrix data in two dimensions.

```python
X, Y = np.meshgrid(month.T, year)

fig, ax = plt.subplots()
ax.contour(X, Y, revenue.T, cmap=cm.coolwarm)
ax.set_xlabel("Month")
ax.set_ylabel("Year")
plt.show()
```
