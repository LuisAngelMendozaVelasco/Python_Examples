# Electricity Usage

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
usage = pd.read_excel("./data/electricity.xlsx", sheet_name='usage', header=None)
monthly = usage.mean(axis=1)
mth = np.array([i for i in range(1, 13)])
```

```python
plt.figure()
plt.bar(mth[[i for i in range(0, len(mth), 3)]], monthly[[i for i in range(0, len(mth), 3)]])
plt.xticks([i for i in range(1, len(mth)+1, 3)])
plt.xlabel("Month")
plt.show()
```

Bars in a bar graph often represent quantities of something best described as text. For instance, each bar of this graph is the electricity usage from the first month of each quarter.
We can use a string array as input to the [xticks()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html#matplotlib-pyplot-xticks) function to modify the x-axis tick labels.

```python
plt.figure()
plt.bar(mth[[i for i in range(0, len(mth), 3)]], monthly[[i for i in range(0, len(mth), 3)]])
plt.xticks(ticks=[i for i in range(1, len(mth)+1, 3)], labels=["January", "April", "July", "October"])
plt.xlabel("Month")
plt.show()
```

The average electricity measurements for all 12 months are in the variable monthly. We will visualize the first months of each quarter in context of the rest of the months.

```python
plt.figure()
plt.bar(mth, monthly)
plt.xticks(ticks=[i for i in range(1, len(mth)+1, 3)], labels=["January", "April", "July", "October"])
plt.xlabel("Month")
plt.show()
```
