# Plot Matrix Columns II

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)
usage = pd.read_excel("./data/electricity.xlsx", sheet_name='usage', header=None)
year = pd.read_excel("./data/electricity.xlsx", sheet_name='yr', header=None)

revenue = price * usage
```

We can provide a vector of x-values when we plot a matrix.

```python
plt.figure()
plt.plot(year.T, revenue.T)
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure()
plt.plot(year.T, revenue.T, ls='--')
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.show()
```
