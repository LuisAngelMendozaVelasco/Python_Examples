# International Gas Prices

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
prices = pd.read_excel("./data/gasprices.xlsx", sheet_name='prices', header=None)
year = pd.read_excel("./data/gasprices.xlsx", sheet_name='yr', header=None)
country = pd.read_excel("./data/gasprices.xlsx", sheet_name='country', header=None)
liter_per_year = pd.read_excel("./data/gasprices.xlsx", sheet_name='litPerYear', header=None)

liter_per_gallon = 3.7854
cost = prices * liter_per_year.to_numpy() / liter_per_gallon
average_cost = cost.mean()
```

* prices: matrix of gasoline prices in USD/gallon where each column represents a different country and each row a different year
* year: vector of year values from 1990 to 2008
* liter_per_gallon: scalar number of liters in a gallon
* liter_per_year: vector containing the average liter consumption of gasoline per individual per year from 1990 to 2008
* country: cell array containing the names of the countries
* cost: annual amount spent (in USD) over the 19 years for a person in each country

```python
plt.figure()
plt.plot(cost)
plt.legend(country.to_numpy().tolist()[0])
plt.show()
```

```python
plt.figure()
plt.bar(x=country.to_numpy().tolist()[0], height=average_cost)
plt.xticks(rotation=90)
plt.ylabel("Gasoline Cost (USD)")
plt.show()
```
