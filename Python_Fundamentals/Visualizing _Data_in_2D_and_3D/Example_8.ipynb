# Plot Electricity Trend

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
residential_electricity = pd.read_csv("./data/resElectricity.csv")

residential = residential_electricity.residential
year = residential_electricity.yr
residential_fitted = residential_electricity.resFitted
```

```python
plt.figure()
plt.plot(year, residential, marker='o', markersize=3, mfc=[0.7, 0.9, 1])
plt.xlabel("Year")
plt.xticks([i for i in range(int(year.min()), int(year.max()), 5)])
plt.ylabel("Residential")
plt.show()
```

```python
plt.figure()
plt.plot(year, residential, marker='o', markersize=3, mfc=[0.7, 0.9, 1])
plt.plot(year, residential_fitted, color='r')
plt.xlabel("Year")
plt.xticks([i for i in range(int(year.min()), int(year.max()), 5)])
plt.ylabel("Residential")
plt.show()
```
