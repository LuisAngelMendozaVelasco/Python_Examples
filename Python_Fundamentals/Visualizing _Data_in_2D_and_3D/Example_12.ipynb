# Monthly and Quarterly Electricity Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)
usage = pd.read_excel("./data/electricity.xlsx", sheet_name='usage', header=None)
year = pd.read_excel("./data/electricity.xlsx", sheet_name='yr', header=None)
month = pd.read_excel("./data/electricity.xlsx", sheet_name='mth', header=None)

revenue = price * usage
```

The variable usage contains electricity usage data for each month from 1990 to 2019.

```python
plt.figure()
plt.plot(month, usage)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure()
plt.plot(year.T, usage.T)
plt.xlabel("Year")
plt.ylabel("Usage")
plt.show()
```

```python
first_month = revenue.iloc[[i for i in range(0, 12, 3)], :]

plt.figure()
plt.plot(year.T, first_month.T)
plt.xlabel("Year")
plt.ylabel("Revenue")
plt.legend(["Jan", "Apr", "Jul", "Oct"])
plt.show()
```
