# Temperatures in Boston

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
temperatures = pd.read_csv("./data/bostonData.csv", header=None)
```

The file bostonData.csv contains the matrix temperatures which contains the daily minimum and maximum temperatures for Boston, Massachusetts in 2005. The first column contains the low temperatures and the second column contains the high temperatures.

```python
plt.figure()
plt.plot(temperatures)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend(["Lows", "Highs"])
plt.show()
```

We can plot a horizontal line using the [axhline()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axhline.html#) function. Similarly, we can plot a vertical line using the [axvline()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axvline.html) function.

```python
low_temperature = temperatures.iloc[:, 0].min()
high_temperature = temperatures.iloc[:, 1].max()

plt.figure()
plt.plot(temperatures)
plt.axhline(low_temperature, color='b')
plt.axhline(high_temperature, color='r')
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.legend(["Lows", "Highs"])
plt.show()
```
