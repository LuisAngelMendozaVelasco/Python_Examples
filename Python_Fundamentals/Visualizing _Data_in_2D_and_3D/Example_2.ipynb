# Scatter plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
gasprices = pd.read_csv("./data/gasprices.csv", header=4)
gasprices.head()
```

```python
Australia = gasprices.Australia
Canada = gasprices.Canada
```

We can use the [scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib-pyplot-scatter) function to create a scatter plot of y against x.

```python
plt.figure()
plt.scatter(Australia, Canada)
plt.xlabel("Australia")
plt.ylabel("Canada")
plt.show()
```

We can change the size of the markers by passing a third input to scatter.

```python
plt.figure()
plt.scatter(Australia, Canada, s=100)
plt.xlabel("Australia")
plt.ylabel("Canada")
plt.show()
```

We can provide a fourth input to scatter to specify the marker color.

```python
plt.figure()
plt.scatter(Australia, Canada, s=50, color='r')
plt.xlabel("Australia")
plt.ylabel("Canada")
plt.show()
```
