# Axis Control

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
gas_prices = pd.read_csv("./data/gasprices.csv", header=4)

Year = gas_prices.Year
Australia = gas_prices.Australia
Germany = gas_prices.Germany
Mexico = gas_prices.Mexico
```

```python
plt.figure()
plt.plot(Year, Australia, ls='--', marker='*')
plt.plot(Year, Germany, ls='--', marker='*')
plt.plot(Year, Mexico, ls='--', marker='*')
plt.xlabel("Year")
plt.xticks([i for i in range(Year.min(), Year.max()+2, 2)])
plt.ylabel("Prices (USD/gal)")
plt.title("Gasoline Prices")
plt.legend(["Australia", "Germany", "Mexico"])
plt.show()
```

We can use the functions [xlim()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlim.html#matplotlib-pyplot-xlim) and [ylim()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylim.html#matplotlib-pyplot-ylim) to set the x- and y-axis limits. Both functions take a vector of two elements as an input.

```python
plt.figure()
plt.plot(Year, Australia, ls='--', marker='*')
plt.plot(Year, Germany, ls='--', marker='*')
plt.plot(Year, Mexico, ls='--', marker='*')
plt.xlabel("Year")
plt.xticks([i for i in range(Year.min(), Year.max()+2, 2)])
plt.ylabel("Prices (USD/gal)")
plt.ylim([0, 8])
plt.title("Gasoline Prices")
plt.legend(["Australia", "Germany", "Mexico"])
plt.show()
```

To focus on the steady increase that started around 2001, we need to change the x- axis limits to start at 2001 and go to the final year.

```python
plt.figure()
plt.plot(Year, Australia, ls='--', marker='*')
plt.plot(Year, Germany, ls='--', marker='*')
plt.plot(Year, Mexico, ls='--', marker='*')
plt.xlabel("Year")
plt.xticks([i for i in range(2001, Year.max()+1)])
plt.xlim([2001, Year.max()])
plt.ylabel("Prices (USD/gal)")
plt.ylim([0, 8])
plt.title("Gasoline Prices")
plt.legend(["Australia", "Germany", "Mexico"])
plt.show()
```

When we change the x- and y-axis limits, the underlying data in the axes is not modified—it's all still there. Only the way the plot is displayed is changed.
If we want to change the axes limits to equal exactly the range of the data, we can use the [autoscale()](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.autoscale.html#matplotlib-axes-axes-autoscale) function.

```python
plt.figure()
plt.plot(Year, Australia, ls='--', marker='*')
plt.plot(Year, Germany, ls='--', marker='*')
plt.plot(Year, Mexico, ls='--', marker='*')
plt.autoscale(enable=True, axis='both', tight=True)
plt.xlabel("Year")
plt.ylabel("Prices (USD/gal)")
plt.title("Gasoline Prices")
plt.legend(["Australia", "Germany", "Mexico"])
plt.show()
```
