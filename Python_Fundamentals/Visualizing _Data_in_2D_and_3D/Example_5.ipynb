# Advanced Annotations

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
r = np.arange(0, 4.1, 0.1)
A = np.pi * r**2
```

```python
plt.figure()
plt.plot(r, A)
plt.xlabel("r")
plt.show()
```

The area of a circle is plotted against the radius. Let's add a y-label with two lines, the first line with the word "Area" and the formula on the second.

```python
plt.figure()
plt.plot(r, A)
plt.xlabel("r")
plt.ylabel("Area \n $\pi r^2$")
plt.show()
```

We can also display variables in labels by using them to create the string label. 

```python
plt.figure()
plt.plot(r, A)
plt.xlabel("r")
plt.ylabel("Area \n $\pi r^2$")
plt.title(f"Area where radius is 0 to {4}")
plt.show()
```
