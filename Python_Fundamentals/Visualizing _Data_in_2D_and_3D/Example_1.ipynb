# stairs, stem, and area

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
gasprices = pd.read_csv("./data/gasprices.csv", header=4)
gasprices.head()
```

```python
Australia = gasprices.Australia
Canada = gasprices.Canada
Year = gasprices.Year

Australia_Canada = Australia - Canada
```

```python
plt.figure()
plt.plot(Year, Australia_Canada)
plt.xlabel("Year")
plt.xticks([i for i in range(1990, 2009, 2)])
plt.title("Difference between Australia and Canada")
plt.show()
```

The syntax for most plotting functions is similar to plot. For example, the [stairs()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stairs.html#matplotlib-pyplot-stairs) function creates a stair plot of yData against xData.

```python
plt.figure()
plt.stairs(Australia_Canada, edges=[i for i in range(1990, 2010)])
plt.xlabel("Year")
plt.xticks([i for i in range(1990, 2009, 2)])
plt.title("Difference between Australia and Canada")
plt.show()
```

[stem()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.stem.html#matplotlib-pyplot-stem) function.

```python
plt.figure()
plt.stem(Year, Australia_Canada)
plt.xlabel("Year")
plt.xticks([i for i in range(1990, 2009, 2)])
plt.title("Difference between Australia and Canada")
plt.show()
```

Area chart using [fill_between()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html#matplotlib-pyplot-fill-between) function.

```python
plt.figure()
plt.fill_between(Year, Australia_Canada)
plt.xlabel("Year")
plt.xticks([i for i in range(1990, 2009, 2)])
plt.title("Difference between Australia and Canada")
plt.show()
```
