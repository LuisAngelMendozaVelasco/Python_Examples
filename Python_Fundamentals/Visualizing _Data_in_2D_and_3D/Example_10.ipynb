# Plot Matrix Columns

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)
usage = pd.read_excel("./data/electricity.xlsx", sheet_name='usage', header=None)

revenue = price * usage
```

The variable revenue contains electricity revenue data for each month from 1990 to 2019. Two-dimensional plotting functions (plot, bar, stem, etc.) behave in the same manner as statistical functions when dealing with matrices: the columns are assumed to be independent variables and plotted as separate lines.

```python
plt.figure()
plt.plot(revenue)
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure()
plt.plot(revenue.T)
plt.xlabel("Years")
plt.ylabel("Revenue")
plt.show()
```

We can use the same line specs for matrices as for vectors.

```python
plt.figure()
plt.plot(revenue, marker='x')
plt.xlabel("Months")
plt.ylabel("Revenue")
plt.show()
```
