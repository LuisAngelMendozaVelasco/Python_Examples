# Plot Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
```

Load data.

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None).to_numpy()
usage = pd.read_excel("./data/electricity.xlsx", sheet_name='usage', header=None).to_numpy()
```

Plot data.

```python
revenue = price * usage
plt.plot(revenue.flatten('F'))
plt.xlim([0, revenue.size])
plt.show()
```
