# Array Creation Functions

```python
import numpy as np
```

The [ones()](https://numpy.org/doc/stable/reference/generated/numpy.ones.html#) function creates an array of given shape, filled with ones.

```python
x = np.ones((3, 4))
x
```

```python
x = 5 * np.ones((3, 4))
x
```

The [rand()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html#numpy-random-rand) function creates an array of uniformly distributed random numbers. Random numbers can be useful for creating simulations.

```python
xU = np.random.rand(100)
xU
```

The [randn()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html#numpy-random-randn) function creates an array of normally distributed random numbers.

```python
xN = np.random.randn(100)
xN
```

The [randint()](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html#numpy-random-randint) function creates a matrix of uniformly distributed random integers.

```python
xI = np.random.randint(10, size=100)
xI
```

We can specify the interval over which the random integers are distributed.

```python
xI2 = np.random.randint(low=11, high=20, size=100)
xI2
```

The uniformly distributed random numbers generated by rand lie between zero and one by default. The rand function doesn't allow us to specify the interval over which the numbers are distributed. If we want uniformly distributed numbers between values a and b, we can compute the following:

x = a + (b - a) * np.random.rand(m, n)

```python
xU2 = -1 + 2 * np.random.rand(100)
xU2
```

Similarly, normally distributed random numbers generated by the randn function will have a mean of zero and a standard deviation of one by default.
If we want normally distributed numbers with mean M and standard deviation sd, you can compute the following: 

x = M + sd * np.random.randn(m, n)

```python
xN2 = 100 + 5 * np.random.randn(100)
xN2
```
