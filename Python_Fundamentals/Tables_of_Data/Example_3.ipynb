# Using sort_values

```python
import pandas as pd
```

```python
EPL = pd.read_csv("./data/EPLresults.csv", dtype={'Team': str})
EPL
```

We can sort a table on a specific variable using the [sort_values()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html#pandas-dataframe-sort-values) function.

```python
winners = EPL.sort_values(by='HomeWins', ignore_index=True)
winners
```

To put the top teams at the top of the table, we need to sort in descending order.

```python
winners = EPL.sort_values(by='HomeWins', ascending=False, ignore_index=True)
winners
```

To sort by a second variables, we can supply them in order to the [sort_values()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html#pandas-dataframe-sort-values) function as a string list.

```python
winners = EPL.sort_values(by=["HomeWins", "AwayWins"], ascending=False, ignore_index=True)
winners
```
