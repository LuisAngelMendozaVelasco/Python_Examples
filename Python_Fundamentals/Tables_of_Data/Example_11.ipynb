# World's Tallest Buildings

```python
import pandas as pd
```

```python
buildings = pd.read_csv("./data/tallest_bldgs.txt", delimiter="\t")
buildings
```

The data in the file tallest_bldgs.txt contains information on the world's tallest buildings as of the year 2010.

```python
buildings["height_m"] = buildings["height_feet"] * 0.3048
```

```python
buildings.sort_values(by="height_feet", ascending=False, inplace=True, ignore_index=True)
buildings
```

```python
five_tallest = buildings[:5]
five_tallest
```

```python
five_tallest.to_csv("./data/tallBldgs.txt", sep="\t")
```
