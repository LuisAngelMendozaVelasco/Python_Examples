# Import EPL Data

```python
import pandas as pd
```

We can use the [read_csv()](https://pandas.pydata.org/docs/reference/api/pandas.read_csv.html#pandas-read-csv) function to import the data from a csv file into a DataFrame.

```python
EPL = pd.read_csv("./data/EPLresults.csv")
EPL
```

There are several ways to store text. Strings (created with double quotation marks) are the easiest to use in most cases. To ensure text is imported as a string, we set the dtype property to string.

```python
EPL = pd.read_csv("./data/EPLresults.csv", dtype={'Team': str})
EPL
```

You can use the [describe()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.describe.html#pandas-dataframe-describe) function to display a summary of each variable.

```python
EPL.describe().applymap('{:.2f}'.format)
```
