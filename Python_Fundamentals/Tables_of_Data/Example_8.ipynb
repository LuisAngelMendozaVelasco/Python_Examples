# Analysis with EPL Table Data

```python
import pandas as pd
```

```python
EPL = pd.read_csv("./data/EPLresults.csv", dtype={'Team': str})
EPL
```

We can add a new variable.

```python
total_wins = EPL.HomeWins + EPL.AwayWins
EPL["TotalWins"] = total_wins
```

```python
EPL["TotalDraws"] = EPL.HomeDraws + EPL.AwayDraws
```

To calculate a team's points, each win is worth three points and each draw worth one point.

```python
EPL["Points"] = 3 * EPL.TotalWins + EPL.TotalDraws
EPL = EPL.sort_values(by="Points", ascending=False, ignore_index=True)
```

```python
EPL.drop(columns=["HomeDraws", "HomeLosses", "HomeWins", "AwayDraws", "AwayLosses", "AwayWins"], inplace=True)
```

You can use the [insert()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.insert.html#pandas-dataframe-insert) and [pop()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.pop.html#pandas-dataframe-pop) functions to move a column to a new location in a table.

```python
EPL.insert(loc=1, column="Points", value=EPL.pop("Points"))
EPL
```
