# Export EPL Table

```python
import pandas as pd
```

```python
EPL = pd.read_csv("./data/EPLtable.csv")
EPL
```

We can use the [to_csv()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.to_csv.html#pandas-dataframe-to-csv) function to create a file from a table.

```python
EPL.to_csv("./data/EPL2015-16.csv")
```

You can specify a delimiter, such as tab, when creating the file.

```python
EPL.to_csv("./data/EPL2015-16.txt", sep="\t")
```
