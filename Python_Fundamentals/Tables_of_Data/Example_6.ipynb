# Subsets of the EPL Data

```python
import pandas as pd
```

```python
EPL = pd.read_csv("./data/EPLresults.csv", dtype={'Team': str})
EPL = EPL.sort_values(by=["HomeWins", "AwayWins"], ascending=False, ignore_index=True)
EPL
```

```python
bottom_goals = EPL.iloc[-3:][["HomeGF", "AwayGF"]]
bottom_goals
```

We can use the [drop()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html#pandas-dataframe-drop) function to remove variables from a DataFrame.

```python
EPL.drop(columns=['HomeGF', 'HomeGA'], inplace=True)
EPL
```

We can use indexing to reorder elements.

```python
EPL = EPL.iloc[:, [0, 1, 4, 2, 5, 3, 6]]
EPL
```
