# Correlation Coefficient of Two Vectors

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
```

```python
data = pd.read_csv("./data/gdp.csv")
data.head()
```

```python
fig, ax1 = plt.subplots()

line_1 = ax1.plot(data.year, data.gdp, marker='*')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xlabel("Year")
ax1.set_ylabel("USD", color='blue')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
line_2 = ax2.plot(data.year, data.medInc, marker='*', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.set_ylabel("USD", color='red')

ax1.legend(line_1 + line_2, ["USA GDP Per Capita", "Median Income in the USA"], loc=0)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

We can use the scatter() function to create a scatter plot of v2 versus v1.

```python
plt.figure()
plt.scatter(data.gdp, data.medInc)
plt.show()
```

We can use the [corr()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html#pandas-dataframe-corr) function to calculate the correlation coefficient between two vectors.

```python
data.iloc[:, 1:].corr()
```
