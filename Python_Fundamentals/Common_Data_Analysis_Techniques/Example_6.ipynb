# Correlation in Electricity Usage

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
```

```python
com = pd.read_excel("./data/electricity.xlsx", sheet_name="com", header=None)
ind = pd.read_excel("./data/electricity.xlsx", sheet_name="ind", header=None)
res = pd.read_excel("./data/electricity.xlsx", sheet_name="res", header=None)

com_v = pd.DataFrame(com.to_numpy().ravel(order='F'))
ind_v = pd.DataFrame(ind.to_numpy().ravel(order='F'))
res_v = pd.DataFrame(res.to_numpy().ravel(order='F'))
usage = ind_v + com_v + res_v
all_sectors = pd.concat([ind_v, com_v, res_v, usage], axis=1)
all_sectors.columns = ["ind", "com", "res", "usage"]
```

```python
plt.figure()
plt.plot(all_sectors)
plt.title("Electricity Usage by Sector")
plt.legend(["Industrial", "Commercial", "Residential"])
plt.show()
```

Correlation between commercial and residential usage. 

```python
all_sectors.iloc[:, 1:3].corr()
```

The DataFrame all_sectors contains the columns of electricity usage data for all of the sectors and the total usage. We can also find the correlation coefficient between them.  

```python
all_sectors.corr()
```
