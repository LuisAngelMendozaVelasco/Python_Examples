# Moving Average

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
x = np.arange(0, 8.1, 0.1)
y = np.sin(x) + 0.1 * np.random.randn(x.size)
```

```python
plt.figure()
plt.plot(x, y, marker='o')
plt.show()
```

Implement moving average as smooth method.

```python
def smooth(a, n):
    """
    a: NumPy 1-D array containing the data to be smoothed
    n: smoothing window size needs, which must be odd number
    """

    out0 = np.convolve(a, np.ones(n, dtype=int), 'valid') / n    
    r = np.arange(1, n-1, 2)
    start = np.cumsum(a[:n-1])[::2] / r
    stop = (np.cumsum(a[:-n:-1])[::2] / r)[::-1]
    
    return np.concatenate((start, out0, stop))

smooth_y = smooth(y, 3)
```

```python
plt.figure()
plt.plot(x, y)
plt.plot(x, smooth_y)
plt.legend(["Input data", "Smoothed data"])
plt.show()
```

We can increase the smoothing window.

```python
smooth_y = smooth(y, 17)

plt.figure()
plt.plot(x, y)
plt.plot(x, smooth_y)
plt.legend(["Input data", "Smoothed data"])
plt.show()
```
