# Visualizing Cricket Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
cricketers = pd.read_csv("./data/cricketers.csv")
cricketers.head()
```

Make a scatter plot of averages

```python
plt.figure()
plt.scatter(cricketers.BowlingAvg, cricketers.BattingAvg, s=10)
plt.plot([20, 65], [20, 65], color="k", ls=':')
plt.xlim([20, 65])
plt.xlabel("Bowling Average")
plt.ylim([20, 65])
plt.ylabel("Batting Average")
plt.show()
```

The table cricketers contains averages for 275 players. The first step is to extract the players that are considered “all-rounders”. A simple criterion for this is any player that has a batting average higher than their bowling average (these players correspond to data points above the dashed line in the scatter plot).

```python
higher_batting = cricketers.BattingAvg > cricketers.BowlingAvg
```

We can use a logical array as an index into an array of any data type.

```python
allround = cricketers[higher_batting]
allround.head()
```

We can specify the size and color of the markers in a scatter plot.

```python
plt.figure()
plt.scatter(cricketers.BowlingAvg, cricketers.BattingAvg, s=10)
plt.scatter(allround.BowlingAvg, allround.BattingAvg, s=10, color='r')
plt.plot([20, 65], [20, 65], color="k", ls=':')
plt.xlim([20, 65])
plt.xlabel("Bowling Average")
plt.ylim([20, 65])
plt.ylabel("Batting Average")
plt.show()
```

To complete the visualization, we will add the names of certain players to the plot, colored according to whether they excel at batting or bowling.

```python
batting = cricketers.BattingAvg > 50
bowling = (cricketers.BowlingAvg < 25) & (cricketers.BattingAvg > 20)
```

You can use the [text()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html#matplotlib-pyplot-text) function to add text to a plot.

```python
plt.figure()
plt.scatter(cricketers.BowlingAvg, cricketers.BattingAvg, s=10)
plt.scatter(allround.BowlingAvg, allround.BattingAvg, s=10, color='r')
for i in batting[batting].index:
    plt.text(cricketers.BowlingAvg[i], cricketers.BattingAvg[i], cricketers.Name[i], color=[0.6, 0.4, 0])
for i in bowling[bowling].index:
    plt.text(cricketers.BowlingAvg[i], cricketers.BattingAvg[i], cricketers.Name[i], color=[0.6, 0, 0.4])
plt.plot([20, 65], [20, 65], color="k", ls=':')
plt.xlim([20, 65])
plt.xlabel("Bowling Average")
plt.ylim([20, 65])
plt.ylabel("Batting Average")
plt.show()
```
