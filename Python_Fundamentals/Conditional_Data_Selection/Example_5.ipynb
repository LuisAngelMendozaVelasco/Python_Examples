# Finding true Values

```python
import pandas as pd
```

```python
results = pd.read_csv("./data/results.csv")
results
```

```python
home_winning = results.homeWins > results.homeLosses
away_winning = results.awayWins > results.awayLosses
home_winning_only = home_winning & ~away_winning
```

The teams are sorted by total season points. The index is therefore equivalent to final standing for the season.

```python
place = home_winning_only[home_winning_only].index
place
```

We can also obtain the locations of just the first or last n true values in a logical vector.

```python
last3_home_winning = home_winning[home_winning].index[-3:]
last3_home_winning
```

```python
last3_away_winning = away_winning[away_winning].index[-3:]
last3_away_winning
```
