# Plotting Electricity Stats by Year and Month

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
com = pd.read_excel("./data/electricity.xlsx", sheet_name='com', header=None)
res = pd.read_excel("./data/electricity.xlsx", sheet_name='res', header=None)
ind = pd.read_excel("./data/electricity.xlsx", sheet_name='ind', header=None)

usage = ind + res + com
year = [i for i in range(1990, 2020)]
```

When we pass an array to the [mean()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.mean.html#pandas-dataframe-mean) function, it calculates the average of each column. The [min()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.min.html) and [max()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.max.html#) functions also calculate the minimum and maximum of the columns of a matrix.

```python
annual = usage.mean()
maximum = usage.max()
minimum = usage.min()

plt.figure()
plt.plot(year, annual)
plt.plot(year, maximum)
plt.plot(year, minimum)
plt.title("Usage")
plt.xlabel("Year")
plt.legend(["Mean", "Max", "Min"])
plt.show()
```

To get monthly statistics, we need to calculate the average, minimum, and maximum of the rows. Many statistical functions accept an optional dimensional argument that specifies whether the operation should be applied to the columns independently (the default) or to the rows.

```python
monthly = usage.mean(axis=1)
max_usage = usage.max(axis=1)
min_usage = usage.min(axis=1)

plt.figure()
plt.bar([i for i in range(1, 13)], monthly)
plt.plot([i for i in range(1, 13)], max_usage, color='r')
plt.plot([i for i in range(1, 13)], min_usage, color='y')
plt.xticks([i for i in range(1, 13)])
plt.show()
```

You can also get the mean, minimum, and maximum, of all of the elements in array.

```python
average_use = usage.mean().mean()
minimum_use = usage.min().min()
maximum_use = usage.max().max()
```
