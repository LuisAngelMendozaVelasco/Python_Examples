# Temperatures in Natick, MA

```python
import pandas as pd
```

```python
temperature_data = pd.read_csv("./data/natickData.csv", header=None)
temperature_data
```

temperature-data is a matrix with the average hourly Fahrenheit temperature in Natick, Massachusetts, home of MathWorks, where each row represents the hour of the day and each column represents the month.

```python
temperature_standard_deviation = temperature_data.std()
temperature_standard_deviation
```

```python
max_standard_deviation = temperature_standard_deviation.max()
max_standard_deviation
```
