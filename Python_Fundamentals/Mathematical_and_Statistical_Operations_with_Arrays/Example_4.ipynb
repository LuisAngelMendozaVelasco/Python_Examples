# Average Electricity Usage in 1990 and 2019

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
com = pd.read_excel("./data/electricity.xlsx", sheet_name='com', header=None)
res = pd.read_excel("./data/electricity.xlsx", sheet_name='res', header=None)
ind = pd.read_excel("./data/electricity.xlsx", sheet_name='ind', header=None)

usage = ind + res + com
first_year = usage.iloc[:, 0]
last_year = usage.iloc[:, -1]
```

```python
plt.figure()
plt.plot([i for i in range(1, 13)], first_year, marker='o', linestyle='dashed')
plt.plot([i for i in range(1, 13)], last_year, marker='o', linestyle='dashed')
plt.xticks([i for i in range(1, 13)])
plt.xlabel("Month")
plt.legend(["First year", "Last year"])
plt.show()
```

The total electricity usage data for 1990 and 2019 have been extracted and saved in firs_year and last_year respectively. We can see in the plot above that the electricity usage in 2019 is higher than in 1990.

```python
average_1990 = first_year.mean()
average_1990
```

```python
average_2019 = last_year.mean()
average_2019
```

There are missing values in last_year. We can identify and count missing values in a vector. To identify missing values in a vector, use the [isna()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isna.html#pandas-dataframe-isna) function.

```python
missing_2019 = last_year.isna()
missing_2019
```
