# Normalizing Electricity Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
com = pd.read_excel("./data/electricity.xlsx", sheet_name='com', header=None)
res = pd.read_excel("./data/electricity.xlsx", sheet_name='res', header=None)
ind = pd.read_excel("./data/electricity.xlsx", sheet_name='ind', header=None)
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)
dollar_2019 = pd.read_excel("./data/electricity.xlsx", sheet_name='dollar2019', header=None)

usage = com + res + ind
price = price * 1000 / 100
revenue = price * usage
```

```python
plt.figure()
plt.plot([year for year in range(1990, 2020)], price.T)
plt.title("Price of Electricity for Each Month ($/MWh)")
plt.xlabel("Year")
plt.show()
```
