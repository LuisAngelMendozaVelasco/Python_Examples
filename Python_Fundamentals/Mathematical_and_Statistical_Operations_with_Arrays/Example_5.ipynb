# Electricity Usage Difference Between 1990 and 2019

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
com = pd.read_excel("./data/electricity.xlsx", sheet_name='com', header=None)
res = pd.read_excel("./data/electricity.xlsx", sheet_name='res', header=None)
ind = pd.read_excel("./data/electricity.xlsx", sheet_name='ind', header=None)

usage = ind + res + com
first_year = usage.iloc[:, 0]
last_year = usage.iloc[:, -1]
usage_difference = last_year - first_year
```

```python
plt.figure()
plt.plot([i for i in range(1, 13)], first_year, marker='o', linestyle='dashed')
plt.plot([i for i in range(1, 13)], last_year, marker='o', linestyle='dashed')
plt.xticks([i for i in range(1, 13)])
plt.xlabel("Month")
plt.legend(["1990", "2019"])
plt.show()
```

We can see in the plot that the electricity usage is higher in 2019 than 1990 for every month. The difference between the electricity data from 2019 and 1990 is saved in the variable usage_difference. We can use the [max()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.max.html#pandas-dataframe-max) function to calculate the highest change in electricity usage.

```python
max_usage, month_number = usage_difference.max(), usage_difference.argmax()
max_usage, month_number
```

The maximum difference occurred in the seventh month, July. To plot the usage for July over the years, we extract the seventh row of usage.

```python
july = usage.iloc[month_number, :]

plt.figure()
plt.plot(july)
plt.title("Usage for July over the years")
plt.xlabel("Years")
plt.xticks(ticks=[i for i in range(1, 31, 5)], labels=[i for i in range(1990, 2020, 5)])
plt.show()
```

We can sort data from largest to smallest using the [sort_values()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sort_values.html#pandas-dataframe-sort-values) function to get the other months with the largest change in electricity usage.

```python
usage_difference.sort_values(ascending=False)
```
