# Electricity Revenue

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
com = pd.read_excel("./data/electricity.xlsx", sheet_name='com', header=None)
res = pd.read_excel("./data/electricity.xlsx", sheet_name='res', header=None)
ind = pd.read_excel("./data/electricity.xlsx", sheet_name='ind', header=None)
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)

usage = com + res + ind
price = price * 1000 / 100
```

```python
plt.figure()
plt.plot(usage)
plt.title("Electricity Usage (MWh)")
plt.xlabel("Month")
plt.show()
```

```python
plt.figure()
plt.plot([year for year in range(1990, 2020)], price.T)
plt.title("Price of Electricity for Each Month ($/MWh)")
plt.xlabel("Year")
plt.show()
```

Electricity revenue for each month from 1990-2019 can be calculated by multiplying each element of price by the respective element of usage using element-wise multiplication.

```python
revenue = price * usage
```

We can round numbers in an array to n decimal places with the [round()](https://numpy.org/doc/stable/reference/generated/numpy.ndarray.round.html#numpy-ndarray-round) function.

```python
revenue_billion = np.ndarray.round(revenue.to_numpy() / 1e9, 3)
```

```python
plt.figure()
plt.plot(price)
plt.title("Monthly Prices for Each Year")
plt.xlabel("Month")
plt.show()
```

As we can see in the plot before, prices have fluctuated between months in a predictable way. So we will calculate monthly inflation rates. To do this, we need to divide the January 2019 price by the price in January 1990, the February 2019 price by February 1990, and so on.

```python
price_inflation = price.iloc[:, -1] / price.iloc[:, 0]
price_inflation
```
