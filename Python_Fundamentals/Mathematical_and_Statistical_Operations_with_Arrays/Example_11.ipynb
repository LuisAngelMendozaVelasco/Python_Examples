# International Gasoline Prices

```python
import pandas as pd
import numpy as np
```

Variables: 
- prices_DPG: matrix of gasoline prices in U.S. dollars per gallon where each column represents a different country and each row a different year
- cx: matrix of conversion values where each column represents a currency corresponding to a country and each row represents the conversion rate for a particular year
- litPerGal: scalar number of liters in a gallon
- litPerYear: vector containing the average liter consumption of gasoline per individual per year from 1990 to 2008

```python
prices_DPG = pd.read_excel("./data/gasprices.xlsx", sheet_name='pricesDPG', header=None)
cx = pd.read_excel("./data/gasprices.xlsx", sheet_name='cx', header=None)
liters_per_year = pd.read_excel("./data/gasprices.xlsx", sheet_name='litPerYear', header=None)

liter_per_gallon = 3.7854
```

To convert gas prices from USD/gallon to a different measurement, we can use the following equation:

$newPrices = oldPrices * otherCurrencyPerUSD / litPerGal$

```python
prices_DPG.head()
```

```python
prices_YPL = prices_DPG.multiply(cx.iloc[:, 5], axis=0) / liter_per_gallon
prices_YPL.head()
```

```python
total_cost = prices_YPL.multiply(liters_per_year[0], axis=0).mean()
total_cost
```
