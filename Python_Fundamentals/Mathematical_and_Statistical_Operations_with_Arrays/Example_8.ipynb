# Weighted Average

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
price = pd.read_excel("./data/electricity.xlsx", sheet_name='price', header=None)
dollar_2019 = pd.read_excel("./data/electricity.xlsx", sheet_name='dollar2019', header=None)
```

```python
plt.figure()
plt.plot(price)
plt.title("Monthly Electricity Prices from 1990 to 2019")
plt.xlabel("Month")
plt.xticks([i for i in range(12)], [i for i in range(1, 13)])
plt.show()
```

Calculating the average monthly prices by calculating the average of each row would not take into account that a dollar in 1990 is worth more than a dollar in 2019. Let's calculate the weighted average of electricity prices from 1990 to 2019, weighted by the value of 1USD relative to 2019.

```python
weight = dollar_2019 / dollar_2019.sum(axis=1)[0]
price_weighted = price.dot(weight.T)
price_weighted
```
