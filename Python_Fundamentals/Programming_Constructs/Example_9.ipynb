# Fibonacci Sequence

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
n = 102
fibonacci = np.ones(n)
```

Create the Fibonacci Sequence.

```python
for k in range(2, n):
    fibonacci[k] = fibonacci[k - 1] + fibonacci[k - 2]
```

```python
plt.figure()
plt.plot(fibonacci)
plt.show()
```
