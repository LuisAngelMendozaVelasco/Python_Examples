# Join Tables

```python
import pandas as pd
```

```python
positions = pd.read_excel("./data/uswnt2019.xlsx", sheet_name='positions')
posTop3 = pd.read_excel("./data/uswnt2019.xlsx", sheet_name='posTop3') 
uswnt = pd.read_excel("./data/uswnt2019.xlsx", sheet_name='uswnt')
uswntTop3 = pd.read_excel("./data/uswnt2019.xlsx", sheet_name='uswntTop3')
```

We can use the join function to merge two DataFrames using any variables with the same names as key variables.

```python
top_players = uswntTop3.join(posTop3.set_index('Player'), on='Player')
top_players
```

```python
uswnt_positions = uswnt.join(positions.set_index('Player'), on='Player')
uswnt_positions
```
