# Concatenating Tables

```python
import pandas as pd
```

```python
EPL = pd.read_excel("./data/EPL.xlsx", sheet_name='EPL')
teamInfo = pd.read_excel("./data/EPL.xlsx", sheet_name='teamInfo') 
games = pd.read_excel("./data/EPL.xlsx", sheet_name='games')
results = pd.read_excel("./data/EPL.xlsx", sheet_name='results', header=None)
```

We can horizontally concatenate two DataFrames, if they have the same number of observations but do not share any variable names.

```python
season = pd.concat([teamInfo, games], axis=1)
season
```

Points and goal difference are stored in the results variable, which doesn't have column names. We can assign new column names with the [columns](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.columns.html#pandas-dataframe-columns) property. Now we can concatenate it with season.

```python
results.columns = ["Points", "GoalDifference"]
season = pd.concat([season, results], axis=1)
season
```
