# Tallest Buildings

```python
import pandas as pd
```

```python
namesYears = pd.read_excel("./data/buildings.xlsx", sheet_name='namesYears')
namesYears.head()
```

```python
buildings = pd.read_excel("./data/buildings.xlsx", sheet_name='buildings')
buildings.head()
```

```python
newBuildings = pd.read_excel("./data/buildings.xlsx", sheet_name='newBuildings')
newBuildings.head()
```

```python
buildings = pd.concat([buildings, newBuildings], axis=0, ignore_index=True)
buildings
```

```python
buildings.rename(columns={'Height_feet_': 'HeightFeet'}, inplace=True)
buildings
```

```python
buildings.sort_values(by="HeightFeet", ascending=False, ignore_index=True, inplace=True)
buildings
```

The table namesYears contains building names and their year of completion.

```python
buildings = buildings.join(namesYears.set_index('BuildingName'), on='BuildingName')
buildings
```
