# Population Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
population_data = pd.read_csv("./data/populationData.csv")
year = population_data.yr
population = population_data.population
```

The vector population contains the estimated world population (in thousands) every other year from 1950 to 2014.

```python
plt.figure()
plt.plot(year, population, marker='o')
plt.show()
```

```python
population[population == 0] = np.nan

plt.figure()
plt.plot(year, population, marker='o')
plt.show()
```

```python
population = population.interpolate()

plt.figure()
plt.plot(year, population, marker='o')
plt.show()
```
