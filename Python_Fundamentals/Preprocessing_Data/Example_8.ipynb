# Fill Missing Values with Unequal Spacing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
data = pd.DataFrame({'x': [0, 2, 3, 4, 7, 9, 10, 12, 15], 'y': [0, np.nan, 7, 8, np.nan, 2, -3, np.nan, -8]})
data
```

```python
plt.figure()
plt.scatter(data['x'], data['y'])
plt.show()
```

By default, the interpolate() function assumes the observations are equally spaced when performing the interpolation.

```python
z = data['y'].interpolate()
z
```

```python
plt.figure()
plt.scatter(data['x'], data['y'])
plt.plot(data['x'], z, marker='x')
plt.show()
```

We can specify the spacing of the observations by using the interpolate index method.

```python
a = data['y'].to_frame().set_index(data['x']).interpolate(method='index')
a
```

```python
plt.figure()
plt.scatter(data['x'], data['y'])
plt.plot(data['x'], z, marker='x')
plt.plot(data['x'], a, marker='x')
plt.show()
```
