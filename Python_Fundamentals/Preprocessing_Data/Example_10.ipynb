# Gasoline Price Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_style("whitegrid")
```

```python
countries = pd.read_excel("./data/gaspriceData.xlsx", sheet_name='countries', header=None, converters={0: str})
prices_raw = pd.read_excel("./data/gaspriceData.xlsx", sheet_name='pricesRaw', header=None)
year_raw = pd.to_datetime(pd.read_excel("./data/gaspriceData.xlsx", sheet_name='yrRaw', header=None, converters={0: str})[0])
```

- pricesRaw: a 19-by-10 matrix of gasoline prices from 10 countries for the years 1990-2008
- yrRaw: a column vector of the years 1990-2008
- countries: a cell array of text containing the countries whose data is in the corresponding columns of pricesRaw

```python
plt.figure()
plt.plot(year_raw, prices_raw)
plt.legend(countries.iloc[0])
plt.show()
```

```python
row_nan = pd.isnull(prices_raw).any(axis=1).to_numpy()
prices = prices_raw.dropna(axis=0)
year = year_raw[~row_nan]
```

```python
plt.figure()
plt.plot(year, prices)
plt.legend(countries.iloc[0], loc=2)
plt.show()
```

The zscore() function will normalize data such that it has a mean of 0 and a standard deviation of 1.

```python
prices_normalized = stats.zscore(prices)
```

```python
plt.figure()
plt.plot(year, prices_normalized)
plt.legend(countries.iloc[0], loc=2)
plt.show()
```
