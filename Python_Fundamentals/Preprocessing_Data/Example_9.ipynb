# Fill Missing Values with Unequal Spacing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
dates = pd.read_excel("./data/electricityData.xlsx", sheet_name="dates", header=None, converters={0: pd.to_datetime})
usage = pd.read_excel("./data/electricityData.xlsx", sheet_name="usage", header=None)
```

```python
usage_filled = pd.DataFrame(usage).interpolate(axis=0)

plt.figure()
plt.plot(dates, usage.iloc[:, 0], marker='.')
plt.scatter(dates.to_numpy(), usage_filled.iloc[:, 0].to_numpy(), marker='x', color='red')
plt.xlim([pd.to_datetime("1995"), pd.to_datetime("2005")])
plt.show()
```

```python
usage_filled_2 = usage.set_index(dates[0]).interpolate(method='index')

plt.figure()
plt.plot(dates, usage.iloc[:, 0], marker='.')
plt.scatter(dates.to_numpy(), usage_filled.iloc[:, 0].to_numpy(), marker='x', color='red')
plt.scatter(dates.to_numpy(), usage_filled_2.iloc[:, 0].to_numpy(), marker='*', color='black')
plt.xlim([pd.to_datetime("1995"), pd.to_datetime("2005")])
plt.show()
```
