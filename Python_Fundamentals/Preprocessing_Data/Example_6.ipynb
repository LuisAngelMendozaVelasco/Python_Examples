# Fill Missing Values

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_style("whitegrid")
```

```python
x = np.array([0, np.nan, 7, 8, np.nan, 2, -3, np.nan, -8])

plt.figure()
plt.plot(x, marker='s')
plt.show()
```

We can use the [fillna()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html) function to use next valid observation to fill gap.

```python
y = pd.DataFrame(x).fillna(method="bfill").to_numpy()

plt.figure()
plt.plot(y, marker='s')
plt.show()
```

We can also fill NaN values using an interpolation method.

```python
z = pd.DataFrame(x).interpolate().to_numpy()

plt.figure()
plt.plot(y, marker='s')
plt.plot(z, marker='s')
plt.show()
```
