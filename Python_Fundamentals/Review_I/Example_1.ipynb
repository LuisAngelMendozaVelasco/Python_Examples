# Women's 200m Backstroke

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set_style("whitegrid")
```

```python
backstroke = pd.read_csv("./data/backstroke.csv")
backstroke.head()
```

The file backstroke.csv contains times for each swimmer from the 2016 Olympic women's backstroke finals. It contains the following three variables.
* time is a matrix containing the times of the eight swimmers (rows) at 50, 100, 150, and 200 meters (columns)
* name is a cell array containing the swimmers' last names
* country is a cell array containing each swimmer's national code

```python
time = pd.concat([pd.DataFrame(np.zeros((backstroke.shape[0], 1))), backstroke[["time_1", "time_2", "time_3", "time_4"]]], axis=1)
d = np.arange(0, 250, 50)

plt.figure()
plt.plot(d, time.T, marker='x')
plt.xlabel("Distance (m)")
plt.ylabel("Time (s)")
plt.legend(backstroke.name)
plt.show()
```

Because all the athletes are so closely matched, it's difficult to see what's happening in this plot. Instead you can see how each swimmer's time compares to the fastest at each distance.

```python
plt.figure()
plt.plot(d, (time - time.min()).T, marker='x')
plt.xlabel("Distance (m)")
plt.ylabel("Time Behind Lead (s)")
plt.legend(backstroke.country)
plt.show()
```

```python
plt.figure()
plt.plot(time.diff(axis=1).T.reset_index(drop=True), marker='x')
plt.xlabel("Lap Number")
plt.ylabel("Split Time (s)")
plt.legend(backstroke.country)
plt.show()
```
