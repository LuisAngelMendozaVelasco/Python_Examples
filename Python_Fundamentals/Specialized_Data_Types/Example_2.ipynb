# Plotting with datetime Variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
EPL = pd.read_csv("./data/EPLtable.csv", encoding='latin-1')
EPL.head()
```

```python
EPL.ManagerHireDate = pd.to_datetime(EPL.ManagerHireDate)
EPL.ManagerHireDate
```

The EPL table contains data about the teams and managers in the English Premier League teams during the 2015-2016 season.

```python
EPL.sort_values(by="ManagerHireDate", inplace=True, ignore_index=True)
EPL
```

```python
plt.figure()
plt.scatter(EPL.ManagerHireDate, EPL.Points, marker='*')
plt.xlabel("Date")
plt.ylabel("Points")
plt.show()
```

Arsène Wenger was the manager of Arsenal for much longer than any of the other managers were managing their teams. We might want to change the x-axis limits to view the cluster of points after 2012.

```python
plt.figure()
plt.scatter(EPL.ManagerHireDate, EPL.Points, marker='*')
plt.xlim((pd.to_datetime("2012-07-01"), pd.to_datetime("2016-07-01")))
plt.xlabel("Date")
plt.ylabel("Points")
plt.show()
```
