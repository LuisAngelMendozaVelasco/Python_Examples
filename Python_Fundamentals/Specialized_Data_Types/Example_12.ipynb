# Tallest Buildings

```python
import pandas as pd
```

```python
names_years = pd.read_csv("./data/names_years.txt", delimiter="\t")
buildings = pd.read_csv("./data/tallest_bldgs.txt", delimiter="\t")
new_buildings = pd.read_csv("./data/new_bldgs.txt", delimiter="\t")

buildings = pd.concat([buildings, new_buildings], axis=0)
buildings.rename(columns={"Height_feet_": "HeightFeet"}, inplace=True)
buildings.sort_values(by="HeightFeet", ascending=False, ignore_index=True, inplace=True)
buildings = buildings.join(names_years.set_index("BuildingName"), on="BuildingName")
buildings
```

```python
buildings.City = pd.Categorical(buildings.City)
buildings.City
```

```python
buildings.City.value_counts()
```

```python
map_dict = {}
for x in buildings.City:
    if x in ["New York City", "NYC"]:
        map_dict[x] = "New York City"
    else:
        map_dict[x] = x

buildings['City'] = buildings['City'].map(map_dict).astype('category')
buildings
```
