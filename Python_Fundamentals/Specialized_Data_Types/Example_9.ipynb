# Premier League Players

```python
import pandas as pd
```

```python
players = pd.read_csv("./data/playerData.csv", encoding='latin-1')
players
```

```python
players.Position = pd.Categorical(players.Position)
players
```

```python
number_striker = (players.Position == "sriker").sum()
number_striker
```

We can view the category names with the number of elements in each category using the [value_counts()](https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html#pandas-series-value-counts) function.

```python
players.Position.value_counts()
```

We can merge multiple categories into one.

```python
map_dict = {}
for x in players.Position:
    if x in ["attacking midfielder", "defensive midfielder", "winger"]:
        map_dict[x] = "midfielder"
    else:
        map_dict[x] = x
players['Position'] = players['Position'].map(map_dict).astype('category')
players
```

```python
map_dict = {}
for x in players.Position:
    if x in ["striker", "forward"]:
        map_dict[x] = "forward"
    else:
        map_dict[x] = x
players['Position'] = players['Position'].map(map_dict).astype('category')
players
```
